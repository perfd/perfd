```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df_plt = df[["tag", "query_latency_mean", "avg_client_server_lat"]].apply(pd.to_numeric, errors='ignore')
with pd.option_context('display.max_rows', None, 'display.max_columns',None):
    display(df_plt.sort_values("tag"))
```

```python
def plot(df, m, ylabel, yticks=None):
    fig = plt.figure(figsize=(16, 2))
    sns.set_style("darkgrid")

    ax = sns.scatterplot(x="tag", y=m, size=8, edgecolor='black', linewidth=0.5, 
                              palette=sns.color_palette("ch:2.5,-.2,dark=.3", n_colors=1), data=df)
    plt.xlabel("")
    plt.xticks(rotation=90)
    if yticks is not None:
        plt.yticks(yticks)
    plt.ylabel(ylabel)
    ax.tick_params(axis='both', which='major', labelsize=12)
    ax.legend().remove()

df_plt_agg = df_plt.groupby("tag").agg({"query_latency_mean": "mean",
                                        "avg_client_server_lat": "mean"}).reset_index()
plot(df_plt_agg, "query_latency_mean", "Avg Qeury Lat. (s)", list(np.arange(7, 10.5, 1.5)))
```
